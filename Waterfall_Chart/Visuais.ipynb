```python
a = ['sales','returns','credit fees','rebates','late charges','shipping']
b = [10,-30,-7.5,-25,95,-7]
```

```python
my_plot = waterfall_chart.plot(a, b)
```

```python
my_plot = waterfall_chart.plot(a, b, rotation_value=90, sorted_value=True, threshold=0.2,
            formatting = "$ {:,.1f}", net_label="End Result", other_label="misc",
            Title="Chart", x_lab = "X axis", y_lab = "Money", blue_color = "blue",
            green_color = "#95FF24", red_color = 'r')
```
